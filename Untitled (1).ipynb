## SFS system

```python
x1 = []
y1 = []
x2 = []
y2 = []
jRx=[]
jRy=[]
HRy=[]

jImx=[]
jImy=[]
HImy=[]
Cond=[]

data= open('CurrentSFS.dat','r')
data1= open('HySFS.dat','r')
data2= open('CondSFS.dat','r')

lines = data.readlines()
lines1 = data1.readlines()
lines2 = data2.readlines()

data.close()
data1.close()
data2.close()
  
for line in lines:
    p=line.split()
    x1.append(float(p[0]))
    y1.append(float(p[1]))
    jRx.append(float(p[2]))
    jRy.append(float(p[3]))
    jImx.append(float(p[4]))
    jImy.append(float(p[5]))

for line in lines1:
    p = line.split()
    x2.append(float(p[0]))
    y2.append(float(p[1]))
    HRy.append(float(p[2]))
    HImy.append(float(p[3]))
    
for line in lines2:
    p = line.split()
   # x2.append(float(p[0]))
   # y2.append(float(p[1]))
    Cond.append(float(p[2]))

yi, xi = np.mgrid[min(y1):max(y1):120j, min(x1):max(x1):300j]
#yi, xi = np.mgrid[-0.5:0.5:20j, -10:10:100j]

############################################

vjRx=griddata((x1,y1), jRx, (xi, yi), method='nearest')
vjRy =griddata((x1,y1), jRy, (xi, yi), method='nearest') 
vjImx=griddata((x1,y1), jImx, (xi, yi), method='nearest')
vjImy =griddata((x1,y1), jImy, (xi, yi), method='nearest') 
vHRy =griddata((x1,y1), HRy, (xi, yi), method='nearest') 
vHImy =griddata((x1,y1), HImy, (xi, yi), method='nearest') 
vCond =griddata((x1,y1), Cond, (xi, yi), method='nearest') 
```

```python
fig=plt.figure()
plt.pcolor(xi,yi,vHImy)
#plt.xlim([0, 40])
#plt.ylim([-4, 4])
fig.suptitle('$Im(H_y)$', fontsize=20)
plt.colorbar()
plt.set_cmap('jet')
plt.show()
```

```python
fig=plt.figure()
plt.pcolor(xi,yi,vHRy)
#plt.xlim([0, 40])
#plt.ylim([-4, 4])
fig.suptitle('$Re(H_y)$', fontsize=20)
plt.colorbar()
plt.set_cmap('jet')
plt.show()
```
